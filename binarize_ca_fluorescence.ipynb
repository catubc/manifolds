```python
####################################
######### BINARIZE DATA ############
####################################
root_dir = '/media/cat/4TB/donato/'
animal_ids = os.listdir(root_dir)
animal_ids = ['DON-002865']

# 
for animal_id in animal_ids:
    
    # 
    sessions = np.sort(os.listdir(os.path.join(root_dir,
                                      animal_id)))
    
    # 
    for session in sessions:
        print (animal_id, "session: ", session)

        fname_out = os.path.join(root_dir,
                     animal_id,
                     session,
                     'suite2p',
                     'plane0',
                     'binarized_traces.npz'
                     )

        if os.path.exists(fname_out):
            continue

        #
        c = calcium.Calcium()
        c.root_dir = root_dir
        c.animal_id = animal_id
        c.session = session

        # 
        c.load_suite2p()

        #
        c.oasis_thresh = 1
        c.F_minwidth = 15
        c.oasis_onphase_minwidth = 4
        c.peak_width = 10

        # 
        c.binarize_fluorescence()

    print ('')
```
